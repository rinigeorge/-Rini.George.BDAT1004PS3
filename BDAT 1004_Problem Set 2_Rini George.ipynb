Step 5. Discover the Male ratio per occupation and sort it from the most to the least.

Step 6. For each occupation, calculate the minimum and maximum ages.

Step 7. For each combination of occupation and sex, calculate the mean age.

Step 8. For each occupation present the percentage of women and men.

### Step 1 - Import the necessary libraries

```python
import pandas as pd
```

### Step 2 - Import the dataset from 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'

### Step 3 - Assign it to a variable called users

```python
# Data are imported and stored in variable named user
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', delimiter='|')
```

### Step 4 - Discover what is the mean age per occupation

```python
# meanage_occupation variable helps the user to display the mean age which is grouped by the occupation
meanage_occupation = users[['occupation', 'age']].groupby('occupation').mean()
meanage_occupation 
```

### Step 5 - Discover the Male ratio per occupation and sort it from the most to the least

```python
# User extract gender and occupation from the data stored in variable 'users'
data = users[['gender', 'occupation']]

# Filter the data of males and store in Total_males. Then it group by occupation and also count it
Total_males = data.loc[data['gender'] == 'M']
GroupBy_occupation = data.groupby('occupation').count()
male_count = Total_males.groupby('occupation').count()

# Male ratio per occupation and sorting it
male_ratio = male_count / GroupBy_occupation
male_ratio.sort_values(ascending=False, by='gender')
```

### Step 6 - For each occupation, calculate the minimum and maximum ages

```python
# Occupation and age details
data = users[['occupation', 'age']]

#Minimum age in each occupation
minimum_age = data.groupby('occupation').min()

#Maximum age in each occupation
maximum_age = data.groupby('occupation').max()

# concating the max and min age according to the occupation
min_max = pd.concat([maximum_age, minimum_age], axis=1)

#printing the final data
min_max
```

### Step 7 - For each combination of occupation and sex, calculate the mean age

```python
# Occupation, gender and age details
data = users[['occupation', 'gender', 'age']]

#Calcualting the mean of the age which is grouped by Occupation and Gender
mean_occuaption_sex = data.groupby(['occupation', 'gender']).mean('age')

#Printing the data
mean_occuaption_sex
```

### Step 8 - For each occupation present the percentage of women and men

```python
#Calculating percentage of women and men for each of the occupation
Percentage = data.groupby('occupation')['gender'].value_counts(normalize=True)*100

#Printing the data
Percentage
```

## Question 6

Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
# Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt

# Reading the data from file and displaying the data
data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
data 
```

```python
# Setting up the figure size
plt.rcParams["figure.figsize"] = [15, 10] 

# Setting the x and y axis in the plot
data.plot( 'Year', ['Marriages_per_1000','Divorces_per_1000'], label=['Marriages', 'Divorces'])

# Mentioning the titles for the plot
plt.title('Marriages and divorces per capita in the U.S. between 1867 and 2014', color = 'g', fontsize = 22)

# Labels - x & y
plt.xlabel('Year', color = 'k', fontsize = 17)
plt.ylabel('Capita', color = 'k', fontsize = 17)

#Displaying the line plot
plt.show()
```
